# Конспект главы 1 «Введение в компьютерные сети и Интернет»
## 1. Что такое Интернет?
**Определение**: Интернет можно описать через аппаратные компоненты или как инфраструктуру, предоставляющую службы для распределенных приложений (веб-серфинг, электронная почта, игры, VoIP и др.).

## 2. Аппаратные компоненты Интернета
* **Хосты** (конечные системы): Устройства, подключающиеся к сети (ПК, смартфоны, серверы, IoT-устройства). Они являются источником и получателем данных.

* **Линии связи**: Каналы передачи данных (медный провод, оптоволокно, радиоканал). Скорость измеряется в битах в секунду (бит/с).

* **Коммутаторы пакетов**:
    * **Маршрутизаторы** (Router): Используются в ядре сети.
    * **Коммутаторы канального уровня** (Switch): Чаще применяются в сетях доступа (LAN).
    * **Функция**: Получают пакет и направляют его по одному из выходных каналов в сторону получателя.

* **Пакет** (Packet): Небольшой сегмент данных, на которые разбивается исходное сообщение для передачи. Содержит заголовок с адресом и служебной информацией.

* **Маршрут** (Path): Последовательность линий связи и коммутаторов, через которые пакет проходит от отправителя к получателю.

* **Поставщики услуг Интернета** (ISP): Компании, предоставляющие конечным пользователям доступ в Интернет. Имеют собственные сети из линий связи и коммутаторов.

* **Протоколы**: Набор правил, управляющих передачей и приемом информации. Главные протоколы — TCP (контроль передачи) и IP (межсетевой протокол). Их совокупность называют стеком TCP/IP.

* **Стандарты**: Разрабатываются IETF (Инженерный совет Интернета) и оформляются в виде документов RFC (Request for Comments). Они детально описывают протоколы (HTTP, SMTP и др.).

## 3. Интернет как инфраструктура для приложений
* **Распределенные приложения**: Выполняются на конечных системах (хостах), обмениваясь данными. Коммутаторы в ядре сети не участвуют в работе приложений, они лишь обеспечивают связь.

* **API Интернета** (сокеты): Программы на конечных системах используют API для отправки/получения данных. Это похоже на правила почтовой службы: нужно правильно упаковать и адресовать данные, чтобы они были доставлены.

## 4. Что такое сетевой протокол?
* **Аналогия с социальным протоколом**: Как и в общении людей (приветствие, вопрос, ответ), сетевой протокол определяет форматы, порядок обмена сообщениями и действия при их передаче/приеме или других событиях.

* **Сетевой протокол**: Управляет взаимодействием аппаратных или программных компонентов устройств. Пример: протокол HTTP для запроса веб-страниц (запрос GET и ответ сервера).

## 5. Периферия сети (Сети доступа и физические среды)
* **Клиенты и серверы**: Хосты делятся на клиентов (запрашивают данные) и серверы (предоставляют данные, часто находятся в дата-центрах).

* **Сеть доступа**: Физически соединяет конечную систему с первым (граничным) маршрутизатором.

* Типы сетей доступа:
    * **DSL**: Использует телефонную линию. Асимметричный доступ (скорость скачивания выше отдачи). Скорость до 24 Мбит/с.
    * **Кабельный** (HFC): Использует сеть кабельного ТВ. Среда разделяемая (скорость падает при множестве активных пользователей). Скорость до 42,8 Мбит/с.
    * **FTTH** (Fiber To The Home): Оптоволокно до квартиры. Может быть AON (активная) или PON (пассивная). Скорость до нескольких Гбит/с.
    * **Ethernet** и **Wi-Fi**: Популярные ЛВС (LAN). Ethernet использует витую пару, Wi-Fi — беспроводную связь. Скорости от 100 Мбит/с до 10 Гбит/с.
    * **Мобильный доступ** (3G, 4G LTE): Использует инфраструктуру сотовой связи. Скорость LTE — более 10 Мбит/с.


* Физические среды передачи:
    * **Витая пара** (медь): Дешевая, распространена в LAN. Скорости 10 Мбит/с – 10 Гбит/с.
    * **Коаксиальный кабель**: Высокие скорости, используется в кабельном ТВ.
    * **Оптоволокно**: Высочайшие скорости (до сотен Гбит/с), низкий уровень помех, подходит для межконтинентальных линий. Дорогое оборудование на коротких дистанциях.
    * **Радиоканалы** (наземные, спутниковые): Не требуют физического кабеля, но подвержены помехам. Спутниковые дают большую задержку (до 280 мс).

Ответы для самопроверки

Предпочтителен FTTH. При 150 сотрудниках 10 Мбит/c ADSL будет недостаточно. Кабельный доступ может сильно замедляться при пиковых нагрузках. Оптоволокно дает стабильно высокую скорость, низкую задержку и большую пропускную способность, необходимую для видеозвонков и облака.

#### Задача 1. «Мой первый сетевой API»: TCP-эхо-сервер и клиент
**Цель**: Понять, как конечная система использует сокеты (API Интернета) для отправки и получения данных, и как работает протокол TCP (соединение, потоковая передача).

Что нужно сделать:
Напишите сервер, который:
1. Слушает на localhost, порт 12345.
2. Принимает входящее TCP-соединение.
3. Читает из него данные до символа перевода строки \n.
4. Отправляет обратно эти же данные (эхо).
5. Закрывает соединение и ждёт следующего.

Напишите клиент, который:
1. Подключается к серверу localhost:12345.
2. Отправляет строку, введённую пользователем с клавиатуры.
3. Получает ответ и выводит его на экран.
4. Закрывает соединение.

Запустите сервер в одном терминале, а клиент — в другом. Проверьте, что строка возвращается.

Иллюстрируемые концепты:

* Хосты (сервер и клиент) — оба являются конечными системами.

* Сокеты как точка входа в API Интернета: socket(), bind(), listen(), accept(), connect(), send(), recv().

* Протокол TCP, управляющий установлением соединения и гарантирующий доставку данных.

* Клиент-серверная модель.



Исходный код

````python
import socket

HOST = '127.0.0.1'
PORT = 12345

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.bind((HOST, PORT))
    s.listen()
    print(f"Сервер слушает на {HOST}:{PORT}")
    while True:
        conn, addr = s.accept()
        with conn:
            print(f"Подключён клиент с адресом: {addr}")
            data = conn.recv(1024)
            if not data:
                break
            conn.sendall(data)
````

**Дополнительное задание**: Модифицируйте сервер так, чтобы он не закрывал соединение после первой строки, а работал, пока клиент не пришлёт пустую строку. Это подчеркнёт понятие «потока данных» в TCP.

* Протоколы IP и ICMP работают совместно для передачи служебной информации.

* Маршрут — вы воочию видите цепочку адресов маршрутизаторов.

* Различие между транспортным (UDP) и сетевым (IP) уровнями.

Я предлагаю задачу на scapy как самый наглядный вариант.

Ключевой фрагмент (для scapy):

````python
from scapy.all import *

target = "8.8.8.8"
for ttl in range(1, 31):
    pkt = IP(dst=target, ttl=ttl) / UDP(dport=33434)
    reply = sr1(pkt, verbose=0, timeout=2)
    if reply is None:
        print(f"{ttl}: *")
    elif reply.type == 11:
        print(f"{ttl}: {reply.src}")
    elif reply.type == 3 and reply.code == 3:
        print(f"{ttl}: {reply.src} (цель достигнута)")
        break
````

* Хосты: браузер обращается к разным серверам в Интернете.
* Стартовый скелет прокси (на основе socketserver):

````python
import socketserver
import socket
import threading

class ProxyHandler(socketserver.BaseRequestHandler):
    def handle(self):
        # self.request – это сокет клиента (браузера)
        data = self.request.recv(4096)
        if not data:
            return
        # Извлекаем первую строку и заголовок Host
        lines = data.split(b'\r\n')
        first_line = lines[0].decode('utf-8', errors='ignore')
        host = ''
        for line in lines[1:]:
            if line.lower().startswith(b'host:'):
                host = line.split(b':', 1)[1].strip().decode('utf-8', errors='ignore')
                break
        print(f"[ПРОКСИ] {first_line} -> Host: {host}")
        # Здесь нужно установить соединение с host:80 и передать данные...
        # (простая реализация оставлена учащемуся)

if __name__ == '__main__':
    with socketserver.ThreadingTCPServer(('localhost', 8888), ProxyHandler) as server:
        print("Прокси запущен на localhost:8888")
        server.serve_forever()
````
